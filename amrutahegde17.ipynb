```python
#6. Draw subplots of pie charts showing % milk productions by cow, buffalo and goat for last 4 years of selected state.
#milkDF["cow"]= milkDF[""]+milkDF["Cow Milk-2013-14"]+milkDF["Cow Milk-2014-15"]+milkDF["Cow Milk-2015-16"]


milkcopy=df.copy()
cow = [milkcopy.get_value(0,"Cow Milk-2011-12"),milkcopy.get_value(0,"Cow Milk-2013-14"),milkcopy.get_value(0,"Cow Milk-2014-15"),milkcopy.get_value(0,"Cow Milk-2015-16")]
boffalo = [milkcopy.get_value(0,"Boffalo Milk-2011-12"),milkcopy.get_value(0,"Boffalo Milk-2013-14"),milkcopy.get_value(0,"Boffalo Milk-2014-15"),milkcopy.get_value(0,"Boffalo Milk-2015-16")]
goat =[milkcopy.get_value(0,"Goat Milk-2011-12"),milkcopy.get_value(0,"Goat Milk-2013-14"),milkcopy.get_value(0,"Goat Milk-2014-15"),milkcopy.get_value(0,"Goat Milk-2015-16")]

fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(wspace = 0.5, hspace = 0.3)

axes[0,0].pie(cow,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[0,0].set_title("cow production")

axes[0,1].pie(boffalo,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[0,1].set_title("Boffalo production")

axes[1,0].pie(goat,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[1,0].set_title("Goat production")
plt.show()
print(boffalo)
```

```python
#milk production question 3

def find_avg(milk_name):
    ans=df.loc[:,"y2011":"y2016"]
    ans=ans.sum(axis=1)/5
    ans=pd.DataFrame(ans)
    ans["State/ UT Name"]=df["State/ UT Name"]
    
    ans=ans[ans["State/ UT Name"]==milk_name]
    
    
    return ans[0].values
print(find_avg("Assam"))
```
